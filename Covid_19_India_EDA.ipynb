```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Importing first dataset and Cleaning it

link : https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset?select=time_series_covid_19_confirmed.csv

```
df=pd.read_csv('/content/datasets_494724_1202764_time_series_covid_19_confirmed.csv')
df.drop(['Lat','Long'],axis=1,inplace=True)
df.head()
```

```
df2=df[df['Country/Region']=='India'].transpose()
df2=df2.iloc[2:,:]
#df2=df2.reset_index()
df2.rename(columns={131:'Cases'},inplace=True)
df2.head()
```

Importing second dataset

link : https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset?select=time_series_covid_19_recovered.csv

```
df3=pd.read_csv('datasets_494724_1202764_time_series_covid_19_recovered.csv')
df3.drop(['Lat','Long'],axis=1,inplace=True)
df4=df3[df3['Country/Region']=='India'].transpose()
df4=df4.iloc[2:,:]
df4.rename(columns={125:'Cases'},inplace=True)
df4.head()
```

Plotting

```
fig,ax=plt.subplots(figsize=(14,7))

ax.plot(df2.index.values,df2.Cases.values,color='red',label='+ confirmed',linewidth=3)
ax.plot(df4.values,color='green',label='recovered',linewidth=3)
#ax.bar(df4.index.values,df2.Cases.values,color='green',label='confirmed')


plt.xticks(range(0, len(df2), 4), df2.index[range(0, len(df2), 4)], rotation = '45',fontsize=12)

plt.xlabel('Month-Day [JAN - MAY]',fontsize=15,fontname = "Times New Roman")
plt.ylabel('Patients',fontsize=15,fontname = "Times New Roman")
plt.title('Covid-19 India [JAN - MAY]',fontsize=20,fontname = "Times New Roman")
plt.legend(loc='best',prop={'size': 15},frameon=True)

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


sns.set_style("darkgrid")  
plt.show()
```
